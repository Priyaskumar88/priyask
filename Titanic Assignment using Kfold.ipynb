```python
y=data1['Survived']  # target
X=data1.drop('Survived',axis=1)  # feature
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X1,y,test_size=0.25,random_state=42) 
```

```python
from sklearn.metrics import confusion_matrix, precision_score,recall_score,f1_score,accuracy_score
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
metric_k=[]    #empty list
neighbors=np.arange(3,15)# taking valuse for 3 to 14 to the variable 

for k in neighbors:   # creating  a loop k (3 to 14)
    # create knn model  where k is 3 first till it is 14 
    knn=KNeighborsClassifier(n_neighbors=k)    # create an instance 
    model1=knn.fit(X_train,y_train)    # create the model
    y_pred_knn=model1.predict(X_test)   # preditction
    acc=accuracy_score(y_test,y_pred_knn)  # accurary
    metric_k.append(acc)   # append accuray to list 
 # check which valuse of k will give highest accuarcy
```

```python
metric_k 
```

```python
plt.plot(neighbors,metric_k,"o-")
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid() # k value is more for 7 so
```

```python
knn=KNeighborsClassifier(n_neighbors=10)
model1=knn.fit(X_train,y_train)
y_pred_knn=model1.predict(X_test)
   
```

```python
confusion_matrix(y_test,y_pred_knn)
```

```python
accuracy_score(y_test,y_pred_knn)
```

```python
recall_score(y_test,y_pred_knn)
```

```python
precision_score(y_test,y_pred_knn)
```

# k fold cross validations

```python
from sklearn.model_selection import KFold
```

```python
kfold_validator=KFold(10)  #  the given dataset and we are splitting into 10 folds and running the Train and Test
```

```python
for train_index,test_index in kfold_validator.split(X1,y):
    print("training index: ",train_index)
    print("testing index: ",test_index)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_result1=cross_val_score(model1,X1,y,cv=kfold_validator) 
cv_result1  #In each iteration, we will get an accuracy score and have to sum them and find the mean.
```

```python
 print(" Average accuracy score KNN after k-fold cross validation : ",np.mean(cv_result1)*100)
```

# statified cross validatoins

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfold_validator=StratifiedKFold(n_splits=10)
```

```python
for train_index,test_index in kfold_validator.split(X1,y):
    print("training index: ",train_index)
    print("testing index: ",test_index)
```

```python
cv_result2 =cross_val_score(model1,X1,y,cv=skfold_validator)
cv_result2
```

```python
 print(" Average accuracy score KNN after stratified k-fold cross validation : ",np.mean(cv_result2)*100)

```
